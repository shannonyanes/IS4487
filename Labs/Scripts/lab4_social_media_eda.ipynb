Dropping Unnessary rows

```python
dfclean= df.drop(columns=['Unnamed: 0', '_id','display_text_range','id', 'in_reply_to_status_id','in_reply_to_user_id','geo','coordinates','contributors','place','scopes','quoted_status','matching_rules','extended_tweet','extended_entities','quoted_status_id'])
```

Change 'created_at' to day date Time format

```python
# Convert the 'date_column' to datetime format
dfclean['created_at'] = pd.to_datetime(df['created_at'], format="%a %b %d %H:%M:%S %z %Y")

# Extract day of the week, time, and date
dfclean['day_of_week'] = dfclean['created_at'].dt.day_name()
dfclean['time'] = dfclean['created_at'].dt.strftime('%H:%M:%S')
dfclean['date'] = dfclean['created_at'].dt.strftime('%Y-%m-%d')


```

## Summarize Data

➡️ Assignment Tasks
- Create at least two charts to summarize and visualize the data
- Examples include: Number of posts per device; top 10 posts with highest engagement

```python
#chart 1
```

Pie Chart to see which day is most popular for people to tweet about the Superbowl

```python
import matplotlib.pyplot as plt
dfclean.groupby('day_of_week').size().plot(kind='pie', autopct='%1.0f%%')
plt.title('Day of the Week People Tweeted')
plt.show()
```

```python
#chart 2
```

Bar chart to see the top 10 Tweeters and how many tweets they tweeted about the commerical

```python
Tweetsbyusers = dfclean.groupby("id_str")["text"].count().sort_values(ascending=False).head(10)
```

```python
fig,axs = plt.subplots(1,1, figsize=(6,3))
Tweetsbyusers.plot(kind='bar', ax=axs)
axs.set_title('Top 10 Users')
axs.set_ylabel("# of Tweets")
plt.show()
```
